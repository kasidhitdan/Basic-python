```python
#Mean imputation
top_penguins = penguins.head(5)
penguins.head(20)
```

```python
avg_value = top_penguins['bill_length_mm'].mean()
avg_value
```

```python
top_penguins= top_penguins['bill_length_mm'].fillna(value=avg_value)
top_penguins
```

```python
#group_by +sum/mean
#penguins[penguins['species']=='Adelie']['bill_length_mm'].mean() #

penguins.groupby('species')['bill_length_mm'].mean()
```

```python
result = penguins.groupby(['island','species'])['bill_length_mm'].agg(['min','mean','median','std','max'])

result.to_csv('result.csv')
```
