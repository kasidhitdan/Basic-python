#**Basic Pandas**

```python
import pandas as pd
```

```python
# create dataframe from scratch

df =pd.DataFrame({
    "name":["Toy","Joe","Marry","John","Anna"],
    "age":[33,25,20,22,31],
    "gender":["M","M","F","M","F"]
})
df
```

```python
df["city"] = ["London","London","Liverpool","Leed","Asenal"]
df
```

```python
df = df.drop("city",axis = 1)
```

```python
df = df.drop(2,axis = 0)
```

```python
df = df.reset_index(drop = True)
df
```

```python
# Rename Columns
list(df.columns)
```

```python
df.columns = ["nickname","age","sex"]
```

```python
df
```

```python
type(df["nickname"])
```

```python
# create a new serie
s1 = pd.Series(["Marry",20,"F"],index = ["nickname","age","sex"])
print(s1)
print(type(s1))
```

```python
df = df.append(s1, ignore_index=True)
df
```

```python
s2 = pd.Series(["London","London","Liverpool","Leed","Asenal"])
df["city"]=s2
df
```

```python
# write csv
df.to_csv("my_data.csv")
```
